```julia
using Plots;pyplot()
using Distributions
using LaTeXStrings

using Zygote
```

## Preapearing the data

```julia
N = 5000                                                          # No of samples

data_1 = rand(MvNormal([0 ; 0], [1 0.75; 0.75 1]), 5000)          # Data Set 1
data_2 = rand(MvNormal([1 ; 4], [1 0.75; 0.75 1]), 5000)          # Data Set 2

X = [data_1  data_2]                                              # Whole data 
labels = [zeros((1,5000)) ones((1,5000))]                         # Lables. 0 for data 1 1 for data 2
```

```julia
# Plotting the data

scatter(data_1[1,:],data_1[2,:],label = L"label = 0",marker = ( 2, 0.3), c= :red)
scatter!(data_2[1,:],data_2[2,:],label = L"label = 1",marker = ( 2, 0.3), c =:green ,leg = :bottomright,size =(800,600))
xlims!(-4,6.5)
lens!([-0.5, 2.0], [1.2, 2.5], inset = (1, bbox(0.7, 0.5, 0.33, 0.25)))
annotate!(5,-1.8,Plots.text("Boundary",:center, 12))
```
